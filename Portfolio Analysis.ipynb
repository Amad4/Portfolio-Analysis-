```python
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import yfinance as yf
```

```python
tickerSymbols = ['TSLA', 'AAPL','MSFT','NVDA','AMZN','AMD','GME']

TSLA = yf.Ticker(tickerSymbols[0])
AAPL = yf.Ticker(tickerSymbols[1])
MSFT = yf.Ticker(tickerSymbols[2])
NVDA = yf.Ticker(tickerSymbols[3])
AMZN = yf.Ticker(tickerSymbols[4])
AMD = yf.Ticker(tickerSymbols[5])
GME = yf.Ticker(tickerSymbols[6])
```

```python
TSLA_df = TSLA.history(period = '3m', start = '2022-07-01', end = '2022-10-01')
AAPL_df = AAPL.history(period = '3m', start = '2022-07-01', end = '2022-10-01')
MSFT_df = MSFT.history(period = '3m', start = '2022-07-01', end = '2022-10-01')
NVDA_df = NVDA.history(period = '3m', start = '2022-07-01', end = '2022-10-01')
AMZN_df = AMZN.history(period = '3m', start = '2022-07-01', end = '2022-10-01')
AMD_df = AMD.history(period = '3m', start = '2022-07-01', end = '2022-10-01')
GME_df = GME.history(period = '3m', start = '2022-07-01', end = '2022-10-01')

```

```python
TSLA = TSLA_df['Close']
AAPL = AAPL_df['Close']
MSFT = MSFT_df['Close']
NVDA = NVDA_df['Close']
AMZN = AMZN_df['Close']
AMD = AMD_df['Close']
GME = GME_df['Close']
```
